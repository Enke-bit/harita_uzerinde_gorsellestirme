- Bu çalışma, dünya ülkelerinin belirli tarihlerdeki sıcaklık verilerini interaktif bir harita üzerinde göstermeyi amaçlamaktadır. Kullanılan veri seti, her ülkenin çeşitli sıcaklık kategorilerindeki (örneğin, 'Aşırı Soğuk', 'Orta Soğuk', 'Orta Sıcak', 'Aşırı Sıcak') sıcaklık verilerini içermektedir.

Projenin adımları şunlardır:

1. Veri Setinin İncelenmesi: İlk olarak, veri seti okunup sütun isimleri ve içeriği incelenmiştir.
Koordinatların Elde Edilmesi: geopy kütüphanesi kullanılarak, her ülkenin enlem ve boylam bilgileri elde edilmiştir.
2. Haritanın Oluşturulması: folium ve MarkerCluster kullanılarak interaktif bir dünya haritası oluşturulmuştur. Haritada her bir ülke, belirli bir sıcaklık değeri ile gösterilmektedir. Kullanıcı bir ülkeye tıkladığında, o ülkenin sıcaklık verileri pop-up olarak görüntülenmektedir.
3. Haritanın Kaydedilmesi: Oluşturulan harita, world_temperature_map.html dosyası olarak kaydedilmiş ve tarayıcıda görüntülenmek üzere hazırlanmıştır.

Bu çalışma, dünya çapında sıcaklık verilerini görselleştirmek ve analiz etmek isteyen araştırmacılar, öğrenciler ve veri bilimciler için yararlı bir araçtır. Harita üzerinde yapılan görselleştirme, ülkeler arası sıcaklık farklarını kolayca gözlemlemeyi ve analiz etmeyi mümkün kılmaktadır.

```python
import pandas as pd

# Veri setini oku
file_path = r'C:\Users\realb\Desktop\asırı sıcaklar tespit modeli/deaths-temperature-gasparrini new.csv'
data = pd.read_csv(file_path)

# Veri setinin ilk birkaç satırını görüntüleyelim
data.head()
```

```python
from geopy.geocoders import Nominatim
import folium
from folium.plugins import MarkerCluster

# Geolocator'ı başlat
geolocator = Nominatim(user_agent="geoapiExercises")

# Ülkelerin koordinatlarını al
def get_coordinates(country):
    try:
        location = geolocator.geocode(country)
        return location.latitude, location.longitude
    except:
        return None, None

# Ülkelerin koordinatlarını ekleyelim
data['Latitude'] = data['country'].apply(lambda x: get_coordinates(x)[0])
data['Longitude'] = data['country'].apply(lambda x: get_coordinates(x)[1])

# Koordinatları eksik olanları filtreleyelim
data = data.dropna(subset=['Latitude', 'Longitude'])

# Haritayı oluştur
m = folium.Map(location=[0, 0], zoom_start=2)
marker_cluster = MarkerCluster().add_to(m)

# Veri setini kullanarak haritaya noktalar ekleyin
for i, row in data.iterrows():
    country = row['country']
    temp = row['Moderate heat']  # Örnek olarak 'Moderate heat' sütununu kullandım
    lat = row['Latitude']
    lon = row['Longitude']
    
    # Pop-up içeriği
    popup_text = f"Ülke: {country}<br>Sıcaklık: {temp}°C"
    
    # Haritaya nokta ekle
    folium.Marker(
        location=[lat, lon],
        popup=popup_text
    ).add_to(marker_cluster)

# Haritayı kaydet
m.save(r'C:\Users\realb\Desktop\asırı sıcaklar tespit modeli/world_temperature_map.html')
```
